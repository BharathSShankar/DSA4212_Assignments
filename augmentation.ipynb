# Data Augmentation using `imgaug`

https://imgaug.readthedocs.io/en/latest/index.html

```python
import numpy as np
import imgaug.augmenters as iaa
import matplotlib.pyplot as plt
```

```python
data_train = np.load("assignment_1_train.npz")
data_val = np.load("assignment_1_val.npz")
```

```python
plt.imshow(data_train["img"][0])
```

```python
# Define augmentations
seq = iaa.Sequential([
    iaa.Fliplr(), # flip images horizontally
])

# Apply the augmentations
data_train_augmented = seq(images=data_train["img"])
```

```python
plt.imshow(data_train_augmented[0])
```

```python
np.savez("assignment_1_train_FlipLR.npz", img=data_train_augmented, label=data_train["label"])
```
